```python
import time
```

# Print the primes till 2000 in multiple ways and measure the time taken to complete

# First way

```python
def first_method_get_prime_numbers(n):
    flag = 0
    i = 2
    while i <= (n/2):
        if (n%i) == 0:
            flag = 1
            break    
        i += 1
    if flag == 0:
        if n < 2:
            pass
        else:
            print(n,' is a prime number.')
```

```python
# Print all prime number till 2000
def first_method_all_prime_number():
    # record start time
    start = time.time()
    for i in range(2000):
        first_method_get_prime_numbers(i)
    # record end time
    end = time.time()
    print("The time of execution of above program is :",
      (end-start) * 10**3, "ms")
 
first_method_all_prime_number()
```

# Second way

```python
def second_method_get_prime_numbers(num):
    if num > 1:
        for i in range(2, (num//2)+1):
            if (num % i) == 0:
                break
        else:
            print(num, "is a prime number")
```

```python
def second_method_all_prime_number():
    # record start time
    start = time.time()
    for i in range(2000):
        second_method_get_prime_numbers(i)
    # record end time
    end = time.time()
    print("The time of execution of above program is :",
      (end-start) * 10**3, "ms")
 
second_method_all_prime_number()
```
